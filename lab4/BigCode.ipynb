```python
import numpy as np

# Створюємо масиви із заданими значеннями
a = np.arange(5, 15)  # Масив із чисел від 5 до 14 (крок за замовчуванням 1)
l = np.linspace(10, 50, num=5)  # Масив із 5 рівновіддалених чисел від 10 до 50

# Виводимо результати
print(f"""
Метод arange дає: {a}
Метод linspace дає: {l}
""")

# Перевіряємо тип елементів у масиві l
print(f"Тип елемента в linspace: {type(l[0])}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Створення трьох підграфіків
fig, (ax1, ax2, ax3) = plt.subplots(1, 3)

# Налаштування заголовків і осей для кожного графіка
ax1.set_title("Досліджуємо arange")
ax1.set_xlabel("X")
ax1.set_ylabel("Значення функції")

ax2.set_title("Досліджуємо linspace")
ax2.set_xlabel("X")
ax2.set_ylabel("Значення функції")

ax3.set_title("Випадкові коливання")
ax3.set_xlabel("X")
ax3.set_ylabel("Рандом")

# Масив з методом arange: змінено діапазон і опис
xa = np.arange(-15, 15, 2)  # Діапазон [-15, 15) із кроком 2
print(f"Для методу arange згенерували: {xa.size} точок")
ya = xa ** 2 * np.sin(xa)  # Інша функція для обчислень

# Масив з методом linspace: збільшено кількість точок і оновлено опис
xl = np.linspace(-15, 15, 150)  # 150 значень між -15 і 15
print(f"Для методу linspace згенерували: {xl.size} точок")
yl = xl ** 2 * np.sin(xl)  # Використовується та сама функція, що й для arange

# Генерація випадкових значень
xr = np.linspace(-15, 15, 50)  # Діапазон для випадкових коливань
yr = np.random.normal(0, 1, 50).round(3)  # Змінено середнє значення й дисперсію
print(f"Згенеровані випадкові значення, max: {yr.max()}, min: {yr.min()}")

# Побудова графіків для arange
ax1.plot(xa, ya, color="purple")  # Змінено колір лінії
ax1.scatter(xa, ya, color='orange', marker='o')  # Змінено маркер і колір

# Побудова графіків для linspace
ax2.plot(xl, yl, color="brown")  # Інший колір лінії
ax2.scatter(xl, yl, color='pink', marker='s')  # Інший маркер

# Побудова графіків для випадкових коливань
ax3.plot(xr, yr, color="blue")
ax3.scatter(xr, yr, color='red', marker='*')  # Змінено маркер і колір

plt.tight_layout()
plt.show()

# Генерація даних для продажів
d = np.arange(1, 31)  # Дні місяця: від 1 до 30
base_sales = np.random.normal(20, 5, size=d.size).round(0)  # Змінено середнє та стандартне відхилення
print("Продажі за місяць:", base_sales)

# Побудова графіка для продажів
plt.plot(d, base_sales, color="green", label="Продажі")
plt.scatter(d, base_sales, color="blue", marker="o")
plt.fill_between(d, base_sales, 20, where=(base_sales > 20), color='orange', label="Продажі успішні")
plt.fill_between(d, base_sales, 20, where=(base_sales < 20), color='red', label="Продажі впали")

# Оформлення графіка
plt.title("Кількість продажів за місяць")
plt.xlabel("День місяця")
plt.ylabel("Кількість проданих товарів")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np

# Генерація масивів із одиницями та нулями різних розмірів
for i in range(5):  # Змінено діапазон для 5 ітерацій
    print(f"Масив із одиниць розміром {i}:")
    print(np.ones(i))
    print(f"Масив із нулів розміром {i}:")
    print(np.zeros(i))
    print("-" * 30)

# Двовимірні масиви із булевими значеннями
ones_matrix = np.ones((5, 5), dtype=bool)  # 5x5 матриця з одиниць
zeros_matrix = np.zeros((5, 5), dtype=bool)  # 5x5 матриця з нулів

print("Двовимірний масив із булевих одиниць:\n", ones_matrix)
print("Двовимірний масив із булевих нулів:\n", zeros_matrix)
```

```python
# Початкова матриця 3x3, заповнена нулями
initial_matrix = np.zeros((4, 4))  # Збільшено розмір до 4x4
print("Початкова матриця (4x4):\n", initial_matrix)

# Ініціалізація вагової матриці 4x4 з різними коефіцієнтами
weights = np.ones((4, 4)) * 0.3  # Коефіцієнт за замовчуванням - 0.3
# Внесення різноманітності у ваги
weights[1, 2] = 1.5  # Центрована вага у другому рядку
weights[0, :] = 1.1  # Верхній рядок
weights[:, 3] = 0.8  # Останній стовпець

print("Вагова матриця:\n", weights)

# Додавання початкової матриці до ваг
result = initial_matrix + weights
print("Результуюча матриця після додавання ваг:\n", result)
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import requests
from io import BytesIO

# URL-адреса зображення
url = "https://images.pexels.com/photos/1695052/pexels-photo-1695052.jpeg?auto=compress&cs=tinysrgb&dpr=1&w=500"

# Завантажуємо зображення з URL
response = requests.get(url)
image = Image.open(BytesIO(response.content))

# Конвертуємо зображення у формат масиву NumPy
image_array = np.array(image)

# Перевіримо форму масиву: наприклад, (висота, ширина, 3) для RGB
print("Shape of image array:", image_array.shape)
print(image_array)

# Інвертуємо кольори зображення (RGB -> 255 - значення)
inverted_image_array = 255 - image_array

# Відображаємо зображення в Jupyter Notebook
plt.figure(figsize=(8, 8))
plt.imshow(inverted_image_array)
plt.axis("off")  # Прибираємо осі для кращого вигляду
plt.show()
```

```python
# Створимо масив ones для коефіцієнта яскравості
brightness_factor = 0.1  # Зменшення яскравості на 50%
filter_matrix = np.ones(image_array.shape) * brightness_factor

# Застосуємо фільтр до зображення (перемножимо поелементно)
darker_image_array = (image_array * filter_matrix).astype(np.uint8)  # Конвертуємо до 8-бітного формату

# Відобразимо зображення
plt.figure(figsize=(8, 8))
plt.imshow(darker_image_array)
plt.axis("off")
plt.show()
```

```python
import numpy as np

X = np.array([[1, 2], [3, 4]])  # Приклад матриці ознак
lambda_reg = 0.1  # Коефіцієнт регуляризації

# Додаємо регуляризаційний терм (λ * одинична матриця)
X_regularized = X + lambda_reg * np.eye(X.shape[0])
print("Регуляризована матриця:\n", X_regularized)
```
